# Spherical p-spin モデルのレプリカ解析
$$ \phi_{\rm 1RSB} (m,q,T) = -\frac{1}{2\beta}
\left \{ \frac{\beta^2}{2} [1+(m-1)q^p] + \frac{m-1}{m} \log (1-q) + \frac{1}{m} \log(1-q+mq) \right　\}
$$
$$\Phi(m, T) =\left \{ m \phi_{\rm 1RSB}(m, q^*, T) \right \} $$
ただし，$q^*$は
$$ \frac{\partial \phi_{\rm 1RSB} (m,q,T)}{\partial q} =0$$
の解．

$\Phi(m, T)$ にもとづいて，$p(=3)$-spin モデルの相図を描く．

```
# 必要なモジュールを読み込む
import numpy as np
from numpy import exp, sqrt, log, pi 
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
```

```
# 関数の定義
def phi1RSB(m, q, T):
    beta = 1/T
    return -1/(2*beta)*(beta**2/2*(1+(m-1)*q**p)+(m-1)/m*log(1-q)+1/m*log(1 + (m-1)*q))

def Phi(m, q, T):
    return m*phi1RSB(m, q, T)

def dPhi_dm(m, q, T):
    beta = 1/T
    return -m/(2*beta)*(beta**2/2*q**p - 1/m**2*log(1+m*q/(1-q))+(1/m)*q/(1-q+m*q)) + phi1RSB(m, q, T)

def qfunc(m, q, T):
    return q-(1-q)*(1-q+m*q)*p/(2*T**2)*q**(p-1)  
```

```
# メインパート
p=3

NT = 700
NTMAX = 1000
Nm = 1000
NmMAX = int(1.4*Nm)
EPS = 1.0e-8

T = np.array([i/NTMAX for i in range(NT)])
ms = np.zeros(T.shape)
mth = np.zeros(T.shape)
mstar = np.zeros(T.shape)

f = -beta/4*np.ones((NT, NmMAX)) # para の自由エネルギーを初期値にした
sigma = np.zeros((NT, NmMAX))
q = np.zeros((NT, NmMAX))

for n in reversed(range(1,NT)): # 高温から解いていく
    beta = 1/T[n] # 逆温度

    m = np.zeros(NmMAX) # m を(0, 1.4) の範囲にNmMAX個設定するための初期化
    q_init = .9 
    for i in reversed(range(NmMAX)):
        m[i] = i/Nm+EPS
        q[n, i], infodict, ier, mesg = fsolve(lambda x: qfunc(m[i], x, T[n]), q_init, full_output = True) # T[n], m[i] に対してq^* を解く
        if abs(infodict['qtf']) > 1.0e-6: # 解けたか否かの判定．解けなかったら解無しと判定し，para解 q^*=0 を代入
            q[n, i] = 0
        q_init = q[n, i] + (q[n, i] < 1.0e-6)*.9 # 得られた解を次の探索の初期値にする．para解になった場合は q_init = .9 を代入

        f[n, i] = dPhi_dm(m[i], q[n, i], T[n]) # f の記録
        sigma[n, i] = m[i]*beta*f[n, i] - beta*Phi(m[i], q[n, i], T[n]) # sigma の記録
    
    if n % 100 == 0: # 100回毎に解をチェック
        plt.subplot(2,1,1)
        plt.plot(m, f[n])
        plt.xlabel('m')
        plt.ylabel('f')

        plt.subplot(2,1,2)
        plt.plot(m, sigma[n])
        plt.xlabel('m')
        plt.ylabel('sigma')
        plt.show()

    if len(m[sigma[n] > EPS]) > 0: # sigma > 0 の解が存在する場合に以下を記録
        mstar[n] = m[sigma[n] > EPS].min() # sigma > 0 になる最小の m 
        mth[n] = m[np.argmax(f[n])] # f が最大になる m 
        ms[n] = m[sigma[n] > EPS].max() # sigma > 0 になる最大の m 
    
    print('T = {} mstar = {} mth = {} ms = {}'.format(T[n], mstar[n], mth[n], ms[n])) # 解のモニタリング
    
```

```
plt.plot(T[ms > EPS], ms[ms > EPS], color = 'b', linewidth = 0.5)
plt.plot(T[mth > EPS], mth[mth > EPS], color ='r', linewidth = 0.5)
plt.plot(T[mstar > 0.011], mstar[mstar > 0.011], color = 'g', linewidth = 0.5)
plt.plot([0, 0.7], [1.0, 1.0], color = 'k', linewidth = 0.5, linestyle = 'dashed')
plt.xlabel('T')
plt.ylabel('m')
plt.show()
```
